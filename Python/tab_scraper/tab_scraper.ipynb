````python
rows = []
meetings = data_json['meetings']

for meet in meetings:

    meeting_name = meet['name']
    meeting_venue = meet['venue']
    meeting_country = meet["country"]
    meeting_code = meet['code']
    meeting_type = meet['type']

    for race in meet['races']:
        race_grade = race['class']
        race_length = race['length']
        race_num = race['number']
        race_time = race['norm_time']
        
        for runner in race['entries']:
            runner_num = runner['number']
            runner_name = runner['name']
            scratched = runner['scratched']

            rows.append([meeting_name, meeting_code,meeting_country, meeting_venue,meeting_type,race_num,race_grade,race_length,race_time,runner_num,runner_name,scratched])

df = pd.DataFrame(rows, columns=['meeting_name', 'meeting_code', 'meeting_country', 'meeting_venue', 'meeting_type', 'race_num', 'race_grade', 'race_length', 'race_time', 'runner_num', 'runner_name', 'scratched'])
df.runner_name = df.runner_name.apply(lambda x: x.replace("'", "").replace(".", "").replace("Res", "").strip().upper())
````

````python
hdr = {'User-Agent':'Mozilla/5.0'}
url = "https://api.beta.tab.com.au/v1/tab-info-service/racing/dates/tomorrow/meetings/?jurisdiction=NSW"
req = urllib.request.Request(url, headers = hdr)
file = urllib.request.urlopen(req)
data = file.read()
data_json = json.loads(data)

meetings = data_json['meetings']
rows = []


for meeting in meetings:
    print(meeting)
    meeting_name = meeting['meetingName']
    location = meeting['location']
    race_type = meeting['raceType']
    # sell_code = meeting['sellCode']['meetingCode']

    

    if race_type!="G":
        continue

    # if location  !="NZL":
    #     continue
    print(meeting)
    try:
        meeting_link = meeting['_links']['races']
        venue_mnemonic = meeting['venueMnemonic']
        meeting_json = json.loads(urllib.request.urlopen(urllib.request.Request(meeting_link, headers = hdr)).read())
        races = meeting_json['races']
        sell_code = meeting['sellCode']['meetingCode']
    except KeyError:
        races = meeting['races']
        venue_mnemonic = meeting_name
        sell_code = "NA"
        meeting_link = "NA"
    # print(meeting_json)
    
    for race in races:
        # print(race)
        race_status = race['raceStatus']
        race_number = race['raceNumber']
        race_name = race['raceName']
        race_time = race['raceStartTime']
        fixed_odds = race['hasFixedOdds']
        race_link = race['_links']['self']

        race_json = json.loads(urllib.request.urlopen(urllib.request.Request(race_link, headers = hdr)).read())

        for runner in race_json['runners']:
            # print(runner)
            # print(race_link)
            runner_name = runner['runnerName']
            runner_number = runner['runnerNumber']
            runner_prop = runner['fixedOdds']['propositionNumber']
            try:
                runner_box = runner['barrierNumber']
            except:
                runner_box = "NA"
            runner_odds = runner['fixedOdds']['returnWin']
            try:
                
                runner_open_odds = runner['fixedOdds']['flucs'][-1]['returnWin']
                runner_delta_change = runner_open_odds-runner_odds
            except KeyError:
                runner_open_odds = runner_odds
                runner_delta_change = 0

            # Step 3: Append the extracted data to the rows list
            rows.append([meeting_name, 
                         location, 
                         sell_code,
                         venue_mnemonic,
                         meeting_link, 
                         race_type, 
                         race_number, 
                         race_status, 
                         race_name, 
                         race_time, 
                         fixed_odds,
                         runner_name,
                         runner_prop,
                         runner_number,
                         runner_box,
                         runner_odds,
                         runner_open_odds,
                         runner_delta_change])

# Step 4: Create a Pandas dataframe from the rows list
df = pd.DataFrame(rows, columns=['meeting_name',
                                  'location',
                                  'sell_code',
                                  'venue_mnemonic',
                                   'meeting_link',
                                   'race_type',
                                    'race_number', 
                                    'race_status', 
                                    'race_name', 
                                    'race_time', 
                                    'fixed_odds',
                                    'runner_name',
                                    'runner_prop',
                                    'runner_number',
                                    'runner_box',
                                    'runner_odds',
                                    'runner_open_odds',
                                    'runner_delta_change'])
````
